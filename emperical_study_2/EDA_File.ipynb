# Feature Engineering

```python
df.head()
```

```python
df=df.drop("id",axis=1)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
#def year_extraction(date):
#         return date.year

#We can do it in the above manner inplace of lambda function, lambda function below does in the same manner as above...

df['year']=df['date'].apply(lambda date: date.year)
df['month']=df['date'].apply(lambda date: date.month)
```

```python
df.head()
```

```python
plt.figure(figsize=(12,8))
plt.subplot().ticklabel_format(useOffset=False,style='plain')
sns.boxplot(x="month",y="price",data=df);
```

```python
plt.figure(figsize=(12,6))
df.groupby('month').mean()['price'].plot();
```

```python
plt.figure(figsize=(12,6))
df.groupby('year').mean()['price'].plot();
```

```python
df=df.drop('date',axis=1)
```

**0 tells that house was not renovated yet...**

```python
df['sqft_basement'].value_counts()
```

```python
X=df.drop('price',axis=1).values
y=df['price'].values

#We used values here because neural networks don't work that well on series and dataframe...
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
!pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model=Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
         batch_size=128,epochs=400)            #We used batches as our data is big...
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot(figsize=(12,6));
```

**There is no overfitting here as the line does not goes up after coming to the mid...**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions=model.predict(X_test)
```

```python
predictions
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
df['price'].describe()
```

**Here we can see that our mean is 5.400881e+05**

**This is the mean price and and the error is about 10000 dollar and we are off by 20%**

```python
explained_variance_score(y_test,predictions)
```
